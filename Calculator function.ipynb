```python
def add(a, b):
    return(a + b)
    
def substract(a, b):
    return (a - b)

def multiply(a, b):
    return (a * b)

def divide(a, b):
    if b == 0:
       return "Error: Cannot divide by zero"
    return (a / b)

#Main program loop
while True:

    print("\n===Accounting Calculator===")
    print("1. Add")
    print("2. Substract")
    print("3. Multiply")
    print("4. Divide")

    choice = input("Choose an operation (1-4): ")
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    if choice == "1":
       print("Result:", add(a, b))
    elif choice == "2":
       print("Result:", substract(a, b))
    elif choice == "3":
       print("Result:", multiply(a, b))
    elif choice == "4":
       print("Result:", divide(a, b))
    else:
         print("Invalid choice")

    again = input("\nDo another calculation? (yes/no)". lower())
    if again != "yes":
        print("Calculator closed")
        break





    
    
```
